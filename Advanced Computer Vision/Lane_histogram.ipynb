```python
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

img = mpimg.imread('lane.jpg')/255

def hist(img):
    bottom_half = img[img.shape[0]//2:,:]
    print("bottom half \n--------------------\n", bottom_half)

    histogram = np.sum(bottom_half, axis=0)
    print("histogram \n--------------------\n", histogram)

    
    return histogram

histogram = hist(img)

plt.plot(histogram)
```
